```python
import matplotlib.pyplot as plt
import pandas as pd
'''
ETH ARB BTC DOGE LINK SOL XRP
'''
symbol_base='ETH'
symbol_quote='USD'
pair=symbol_base+'-'+symbol_quote

# Sample data
data=pd.read_csv('/Users/andyma/Desktop/Python /Tokkalabs_WorkSpace/RabbitX/RabbitX_FundingRate_'+pair+'.csv',index_col=0)

# Convert 'A' to datetime format
data['time_hour'] = pd.to_datetime(data['time_hour'])
data['funding_rate_apy'] = data['funding_rate_apy'].interpolate(method='linear')
data['Binance'] = data['Binance'].interpolate(method='linear')
data['Bybit'] = data['Bybit'].interpolate(method='linear')
data['OKX'] = data['OKX'].interpolate(method='linear')
data['dYdX'] = data['dYdX'].interpolate(method='linear')
data['Synthetix'] = data['Synthetix'].interpolate(method='linear')
# Plot the data
plt.figure(figsize=(15, 4),dpi=300)

# Plotting each column as a line
plt.plot(data['time_hour'], data['funding_rate_apy'], label='RabbitX')
plt.plot(data['time_hour'], data['Binance'], label='Binance',lw=0.9)
plt.plot(data['time_hour'], data['Bybit'], label='Bybit',lw=0.9)
plt.plot(data['time_hour'], data['OKX'], label='OKX',lw=0.9)
plt.plot(data['time_hour'], data['dYdX'], label='dYdX',lw=0.9)
plt.plot(data['time_hour'], data['Synthetix'], label='Synthetix',lw=0.9)

# Adding labels and title
plt.xlabel('Date')
plt.ylabel('Funding Rate (APY)')
plt.title(pair+' Funding Rate (APY)')
plt.legend()

# Set the y-axis limit
plt.ylim(top=4)

# Show the plot
plt.grid(True)

plt.savefig('/Users/andyma/Desktop/Python /Tokkalabs_WorkSpace/RabbitX/FundingRate_volume_withCEX_plots/'+pair+'.png')
print('finished')
plt.show()
```

```python
import pandas as pd
from datetime import datetime
import numpy as np
import plotly.graph_objects as go
from collections import defaultdict
import plotly

asset = 'DOGE'

df= pd.read_csv('/Users/andyma/Desktop/Python /Tokkalabs_WorkSpace/RabbitX/Synthetix_fundingRate_graph_candel_'+asset+'.csv',index_col=0)
# Plot candlestick chart
# Determine colors based on whether close > open
colors = []
for i in range(len(df)):
    if df.loc[i, 'close'] > df.loc[i, 'open']:
        colors.append('green')
    else:
        colors.append('red')

# Plot candlestick chart
fig = go.Figure(data=[go.Candlestick(
    x=df['x'],
    open=df['open'],
    high=df['high'],
    low=df['low'],
    close=df['close'],
    increasing=dict(line=dict(color='green')),
    decreasing=dict(line=dict(color='red')),
)])

# Update layout
fig.update_layout(
    title='Hourly Candlestick Chart of Funding Rates ('+asset+'-USD)',
    xaxis_title='Date',
    yaxis_title='Funding Rate (APY)',
    xaxis_rangeslider_visible=False,
    yaxis=dict(range=[0, 4]),
    width=1300,  # Setting width of the figure
    height=500,  # Setting height of the figure
)


# # Save as PNG with specified DPI
# fig.write_image('/Users/andyma/Desktop/Python /Tokkalabs_WorkSpace/RabbitXFundingRate_volume_withCEX_plots/Synthetix_fundingRate_graph_candel_'+asset+'.png', engine='kaleido', scale=5)  # DPI = 300, scale=5 to match figsize
plotly.offline.plot(fig, filename='../lifeExp.html')
# Show the plot (optional)
fig.show()
```
