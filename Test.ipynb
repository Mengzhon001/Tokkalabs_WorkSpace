```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
days = np.arange(1, 11)  # Days from 1 to 10
income = [100, 200, 150, 300, 250, 100, 400, 300, 200, 150]
expenses = [50, 100, 75, 200, 150, 60, 250, 200, 100, 90]
interest_rates = [1.5, 1.6, 1.7, 1.8, 1.5, 1.4, 1.6, 1.7, 1.8, 1.9]  # Interest rates for 10 days

# Convert expenses to negative values
expenses = [-e for e in expenses]

# Create the bar chart
fig, ax1 = plt.subplots()

# Plot income as bars
bars_income = ax1.bar(days, income, width=0.4, label='Income', color='g')

# Plot expenses as bars
bars_expenses = ax1.bar(days, expenses, width=0.4, label='Expenses', color='r')

# Add labels, title, and legend for the bar chart
ax1.set_xlabel('Days')
ax1.set_ylabel('Amount')
ax1.set_title('Income, Expenses, and Interest Rates over 10 Days')
ax1.legend(loc='upper left')

# Create a second y-axis for the interest rates
ax2 = ax1.twinx()
ax2.set_ylabel('Interest Rate (%)')
line_interest = ax2.plot(days, interest_rates, label='Interest Rate', color='b', marker='o')

# Combine legends from both y-axes
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='upper center')

# Optionally, add the values on top of the bars
def add_labels(bars, ax):
    for bar in bars:
        height = bar.get_height()
        ax.annotate('{}'.format(height),
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

add_labels(bars_income, ax1)
add_labels(bars_expenses, ax1)

# Display the chart
plt.show()
```
